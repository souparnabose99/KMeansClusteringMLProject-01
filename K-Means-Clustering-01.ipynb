# Project demonstrating K-Means-Clustering Algorithm

<b>Import required Libraries<b/>

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

<b>Generate artificial data using Blobs<b/>

```python
from sklearn.datasets import make_blobs
```

```python
blob_data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
plt.scatter(blob_data[0][:, 0], blob_data[0][:, 1], c=blob_data[1], cmap='rainbow')
```

<b>Using K-Means algorithm<b/>

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(blob_data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

<b>Plotting the data to compare Algorithm vs Original for 4 cluster<b/>

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize = (10,6))

ax1.set_title('K-Means')
ax1.scatter(blob_data[0][:,0], blob_data[0][:, 1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(blob_data[0][:,0], blob_data[0][:, 1], c=blob_data[1], cmap='rainbow')
```

<b>Plotting the data to compare Algorithm vs Original for 3 cluster<b/>

```python
kmeans_2 = KMeans(n_clusters=3)
kmeans_2.fit(blob_data[0])
kmeans_2.labels_

fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize = (10,6))

ax1.set_title('K-Means')
ax1.scatter(blob_data[0][:,0], blob_data[0][:, 1], c=kmeans_2.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(blob_data[0][:,0], blob_data[0][:, 1], c=blob_data[1], cmap='rainbow')
```

<b>Plotting the data to compare Algorithm vs Original for 2 cluster<b/>

```python
kmeans_3 = KMeans(n_clusters=2)
kmeans_3.fit(blob_data[0])
kmeans_3.labels_

fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize = (10,6))

ax1.set_title('K-Means')
ax1.scatter(blob_data[0][:,0], blob_data[0][:, 1], c=kmeans_3.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(blob_data[0][:,0], blob_data[0][:, 1], c=blob_data[1], cmap='rainbow')
```
